```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
x = np.arange(-6.0, 6.0, 0.1)
```

```python
y = sigmoid(x)
```

```python
plt.plot(x, y)
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
y_relu =relu(x)
```

```python
plt.plot(x, y_relu)
```
